# Analyze data to know the insight of two questions below

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cấu hình giao diện biểu đồ cho đẹp
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## Read the dataset

```python
df = pd.read_csv('../../data/student_academic_performance_1M.csv')

df.head()
```

## Question 1: Việc sử dụng AI và thời gian thực hành ảnh hưởng như thế nào đến điểm môn học kỹ thuật và điểm trung bình tổng

### 1.1 Kiểm tra phân phối của đặc trưng thể hiện thời gian thực hành `coding_practice_hours` của sinh viên

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='coding_practice_hours', bins=30, kde=True, color='teal')
plt.title('Phân phối thời gian thực hành Code (Coding Practice Hours)')
plt.xlabel('Giờ thực hành')
plt.ylabel('Số lượng sinh viên')
plt.show()
```

- Biểu đồ hình chuông và lệch phải, trong số lượng sinh viên không thực hành chiếm đa số

### 1.2 Kiểm tra phân bố của đặc trưng sử dụng AI `ai_tool_usage` của học sinh

```python
# Vẽ biểu đồ đếm (Count Plot)
# Lưu ý: thay 'df' bằng tên biến dataframe của bạn
ax = sns.countplot(x='ai_tool_usage', data=df, palette=['#95a5a6', '#3498db'])

# Trang trí tiêu đề và nhãn
plt.title('Số lượng Sinh viên: Không dùng AI vs Có dùng AI', fontsize=15, fontweight='bold')
plt.xlabel('Trạng thái sử dụng', fontsize=12)
plt.ylabel('Số lượng sinh viên', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Không sử dụng (0)', 'Có sử dụng (1)'])

# Thêm con số cụ thể trên đầu cột (Data Labels)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 8), textcoords='offset points',
                fontsize=12, fontweight='bold')

plt.show()
```

- Số lượng học sinh không sử dụng AI chiếm áp đảo so với học sinh có sử dụng

### 1.3 Việc sử dụng AI tác động đến điểm môn học kỹ thuật (`computer_score`) và điểm trung bình tổng (``final_gpa) như thế nào

```python
# Tạo Box Plot so sánh
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Biểu đồ 1: AI vs Computer Score
sns.boxplot(ax=axes[0], data=df, x='ai_tool_usage', y='computer_score', palette="Set2")
axes[0].set_title('Tác động của AI đến Điểm môn Máy tính')
axes[0].set_xticklabels(['Không dùng (0)', 'Có dùng (1)'])

# Biểu đồ 2: AI vs GPA
sns.boxplot(ax=axes[1], data=df, x='ai_tool_usage', y='final_gpa', palette="Set2")
axes[1].set_title('Tác động của AI đến GPA Tổng kết')
axes[1].set_xticklabels(['Không dùng (0)', 'Có dùng (1)'])

plt.tight_layout()
plt.show()
```

### 1.3 Việc sử dụng AI kết hợp với thời gian học sinh thực hành tác động như thế nào đến điểm số

```python
# BƯỚC XỬ LÝ: Làm tròn giờ code để gom nhóm (Binning)
# Việc này giúp biểu đồ mượt hơn và giảm nhiễu
df['coding_hours_rounded'] = df['coding_practice_hours'].round().astype(int)

plt.figure(figsize=(12, 6))

# Vẽ Line Plot với confidence interval (vùng mờ xung quanh đường)
# hue='ai_tool_usage': Tách thành 2 đường màu khác nhau
sns.lineplot(
    data=df, 
    x='coding_hours_rounded', 
    y='computer_score', 
    hue='ai_tool_usage', 
    palette=['grey', 'blue'],  # Màu xám cho ko dùng, Xanh cho có dùng
    marker='o',                # Đánh dấu điểm dữ liệu
    style='ai_tool_usage',     # Kiểu đường (nét liền/đứt) khác nhau
    markers=True, dashes=False
)

plt.title('Mối quan hệ: Giờ Code - Điểm Số - Sử dụng AI', fontsize=15)
plt.xlabel('Thời gian Code (Đã làm tròn theo giờ)', fontsize=12)
plt.ylabel('Điểm môn Máy tính trung bình', fontsize=12)
plt.legend(title='Sử dụng AI', labels=['Không (0)', 'Có (1)'])
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
